This is a weekly pattern summary task.
I should avoid a generic dashboard-style summary.
I should rank the most repeated instability patterns across days and meals, then point out which one is most actionable.
Daily-level and meal-level tables should be enough; raw CGM is not needed unless the higher-level summaries are missing.

```python
daily_sub = daily_df.copy()
meal_sub = meal_features_df.copy()

summary_lines = []

if "num_spikes" in daily_sub.columns and "date" in daily_sub.columns:
    worst_days = daily_sub.sort_values("num_spikes", ascending=False).head(3)[["date", "num_spikes"]]
    summary_lines.append("Top spike-heavy days:")
    for _, row in worst_days.iterrows():
        summary_lines.append(f"- {row['date']}: {row['num_spikes']} spikes")

if "meal_type" in meal_sub.columns and "peak_delta" in meal_sub.columns:
    meal_rank = (
        meal_sub.groupby("meal_type")["peak_delta"]
        .agg(["mean", "count"])
        .sort_values("mean", ascending=False)
        .head(3)
    )
    summary_lines.append("Meal types ranked by average spike size:")
    for idx, row in meal_rank.iterrows():
        summary_lines.append(f"- {idx}: mean peak_delta={row['mean']:.1f}, n={int(row['count'])}")

if "cv_glucose" in daily_sub.columns:
    unstable_days = int((daily_sub["cv_glucose"] >= 0.25).sum())
    summary_lines.append(f"Days with elevated variability (cv_glucose >= 0.25): {unstable_days}")

print("\n".join(summary_lines))
```
